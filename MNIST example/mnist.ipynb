```python
# Importar o dataset e as bibliotecas necessárias 
from tensorflow.keras.datasets import mnist
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# Carregar os dados 
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
# Criar o modelo sequencial
model = keras.Sequential([
    layers.Dense(512, activation='relu'), # units=512
    layers.Dense(10, activation='softmax') # units=10
])
```

```python
# Compilar o modelo 
model.compile(
    optimizer='rmsprop',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Processar e normalizar os dados 
train_images = train_images.reshape((60000, 28*28))
train_images = train_images.astype('float32')/255
test_images = test_images.reshape((10000,28*28))
test_images = test_images.astype('float32')/255
```

```python
# Treinar o modelo 
model.fit(
    train_images, train_labels, epochs=5, batch_size=128
)
```

```python
# Avaliar modelo com dados de teste 
test_loss, test_acc = model.evaluate(test_images, test_labels)
```
