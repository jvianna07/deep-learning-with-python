```python
from tensorflow.keras.datasets import mnist

from tensorflow import keras
from tensorflow.keras import layers
```

```python
(train_images, train_labels), (test_labels, test_labels) = mnist.load_data()
```

```python
model = keras.Sequential([
    layers.Dense(512, activation='relu'), # units=512
    layers.Dense(10, activation='softmax') # units=10
])
```

```python
model.compile(
    optimizer='rmsprop',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```
